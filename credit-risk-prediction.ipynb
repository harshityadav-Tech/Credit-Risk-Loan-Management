```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("loan_data_200.csv")
fig, axs = plt.subplots(2, 2, figsize=(14,10))
fig.suptitle("Loan Applicant Data Visualizations", fontsize=16, fontweight="bold")


axs[0,0].hist(df["Loan_Amount"], bins=20, color="skyblue", edgecolor="black")
axs[0,0].set_title("Distribution of Loan Amount")
axs[0,0].set_xlabel("Loan Amount")
axs[0,0].set_ylabel("Number of Applicants")


colors = df["Default"].map({0:"lightblue",1:"yellow"})
axs[0,1].scatter(df["Income"], df["Loan_Amount"], c=colors, alpha=0.6)
axs[0,1].set_title("Income vs Loan Amount (Default vs No Default)")
axs[0,1].set_xlabel("Income")
axs[0,1].set_ylabel("Loan Amount")


credit_default = df.groupby("Credit_History")["Default"].sum()
axs[1,0].bar(["Bad Credit (0)", "Good Credit (1)"], credit_default, color=["lightblue","yellow"])
axs[1,0].set_title("Number of Defaults by Credit History")
axs[1,0].set_ylabel("Number of Defaults")


default_counts = df["Default"].value_counts()
axs[1,1].pie(default_counts,
             labels=["No Default","Default"],
             autopct="%1.1f%%",
             colors=["lightblue","yellow"],
             startangle=90)

axs[1,1].set_title("Proportion of Default vs Non-Default")

plt.tight_layout()
plt.show()
```
